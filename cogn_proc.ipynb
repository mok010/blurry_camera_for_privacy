```python
#기본적인 라이브러리와 데이터 불러오기
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

train_target = pd.read_csv("cogn_data/training_label.csv")
test_target = pd.read_csv("cogn_data/val_label.csv")
train_input = pd.read_csv("cogn_data/train_mmse.csv")
test_input = pd.read_csv("cogn_data/val_mmse.csv")
```

```python
#EMAIL에 대한 라벨링을 EMAIL로 통일
```

```python
train_target = train_target.rename(columns={'SAMPLE_EMAIL':'EMAIL'})
train_input = train_input.rename(columns={'SAMPLE_EMAIL':'EMAIL'})
train_target
```

```python
#데이터를 쉽게 구분하기 위해 이메일에서 숫자만 제거
```

```python
train_input
```

```python
train_input['EMAIL'] = train_input['EMAIL'].str.replace('nia+','')
train_input['EMAIL'] = train_input['EMAIL'].str.replace('@rowan.kr','')
train_target['EMAIL'] = train_target['EMAIL'].str.replace('nia+','')
train_target['EMAIL'] = train_target['EMAIL'].str.replace('@rowan.kr','')
```

```python
#정렬 및 병합을 위해 EMAIL을 object형에서 int형으로 변경
```

```python
train_target = train_target.astype({'EMAIL':int})
train_input = train_input.astype({'EMAIL':int})
train_input.dtypes
train_target.dtypes
```

```python
#병합을 위해 정렬
```

```python
train_input = train_input.sort_values('EMAIL')
train_target = train_target.sort_values('EMAIL')
```

```python
#병합을 위해 DataFrame형식으로 변경
```

```python
train_input = pd.DataFrame(train_input)
train_target = pd.DataFrame(train_target)
```

```python
#각각의 병도에 매칭시켜 이메일을 병도로 변경
```

```python
train_set = pd.merge_asof(train_input, train_target, on='EMAIL')
train_set
```

```python
#이메일에서 병도로 바꾼 열을 target data로 분리
```

```python
train_set
```

```python
train_target = train_set['DIAG_NM_x']
train_target
```

```python
#input data에 필요없는 'DIAG_NM' , 'EMAIL' 열을 삭제
```

```python
train_input = train_set.drop(['DIAG_NM_x', 'EMAIL'], axis=1)
train_input
```

```python
#test data들도 train data와 같이 처리
```

```python
test_target = test_target.rename(columns={'SAMPLE_EMAIL':'EMAIL'})
test_input = test_input.rename(columns={'SAMPLE_EMAIL':'EMAIL'})
test_target
test_input['EMAIL'] = test_input['EMAIL'].str.replace('nia+','')
test_input['EMAIL'] = test_input['EMAIL'].str.replace('@rowan.kr','')
test_target['EMAIL'] = test_target['EMAIL'].str.replace('nia+','')
test_target['EMAIL'] = test_target['EMAIL'].str.replace('@rowan.kr','')
test_target = test_target.astype({'EMAIL':int})
test_input = test_input.astype({'EMAIL':int})
test_input.dtypes
test_target.dtypes
test_input = test_input.sort_values('EMAIL')
test_target = test_target.sort_values('EMAIL')
test_input = pd.DataFrame(test_input)
test_target = pd.DataFrame(test_target)
test_set = pd.merge_asof(test_input, test_target, on='EMAIL')
test_set
test_target = test_set['DIAG_NM_x']
test_target
test_input = test_set.drop(['DIAG_NM_x', 'EMAIL'], axis=1)
test_input
```

```python
#현재는 처리방법을 찾지 못한 object 변수들을 제거 -> 후에 처리방법을 알아봐도 좋을듯함
```

```python
test_input.dtypes
```

```python
test_input = test_input.drop(['DOCTOR_NM', 'DIAG_NM_y'], axis=1)
test_input.dtypes
```

```python
test_target.dtypes
```

```python
train_target.dtypes
```

```python
train_input = train_input.drop(['DOCTOR_NM', 'DIAG_NM_y'], axis=1)
train_input.dtypes
```

```python
#로지스틱 회귀 모델 적용 (85% 75%로 다른 모델들 보다야 훨씬 나은데 좋은 모델이라면 둘다 90퍼 대로 나오는게 좋다)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr = LogisticRegression(C=20, max_iter=1000)
lr.fit(train_input, train_target)

print(lr.score(train_input, train_target))
print(lr.score(test_input, test_target))
```

```python
#경사하강법분류모델도 정확도가 아직 높지않음
```

```python
from sklearn.linear_model import SGDClassifier
```

```python
sc = SGDClassifier(loss='log_loss', max_iter=10, random_state=42)
sc.fit(train_input, train_target)

print(sc.score(train_input, train_target))
print(sc.score(test_input, test_target))
```
